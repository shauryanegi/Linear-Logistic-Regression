### Gradient Descent

```python
x = -8
lr = 0.05

error_list = []

x_list = []

for i in range(100):
    grad = 2 * (x - 5)

    #Updation Rule
    x = x - lr * grad
    x_list.append(x)

    e = (x - 5)**2
    error_list.append(e)
```

```python
x
```

```python
error_list
```

```python
plt.plot(error_list)
plt.xlabel('Iterations')
plt.ylabel('Loss value')
plt.show()
```

```python
plt.plot(X, err)
plt.scatter(x_list, error_list)
plt.show()
```
